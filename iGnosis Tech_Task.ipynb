```python
# Most profitable customer >> assuming that profit on all products is same. 
# Most profitable customer >> the customer who generates high money (TOT_SALES) >> They can also be loyal customers

# As per loyalty card no.
df.groupby(['LYLTY_CARD_NBR', 'PREMIUM_CUSTOMER', 'LIFESTAGE'])['TOT_SALES'].sum().nlargest(10)
```

### Insight:

#####  Our most profitable customers are the mostly the Budget friendly OLDER FAMILIES

```python
# top 3 most bestselling/profitable products

df.groupby(['PROD_NBR', 'PROD_NAME'])['TOT_SALES'].sum().nlargest(3)
```

```python
# characteristics of the most frequent customers type

df.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['PROD_QTY'].sum().nlargest(10)
```

The most loyal customer segment appears to be Older Families in the Budget category, who have spent the most on purchases. 
This suggests that they prioritize affordability and likely look for value-for-money products.
